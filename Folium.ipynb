```python
def adiciona_cidade(m,inclusao,color,icon,comment):
    df = pd.read_csv('worldcities.csv')

    for i in inclusao:
        nome = i
        lat = df.loc[(df['city']==i)&(df['country']=='Germany'),'lat'].to_list()[0]
        lng = df.loc[(df['city']==i)&(df['country']=='Germany'),'lng'].to_list()[0]
        aux = {'nome':nome,'lat':lat,'lng':lng}
        
        folium.Marker([aux['lat'],aux['lng']],
                      popup=comment,
                      icon = folium.Icon(color=color,
                                        icon_color='white',
                                        prefix='fa',
                                        icon=icon)).add_to(m)

        
def adiciona_circulo(m,inclusao,km,color):
    km = 1000*km
    df = pd.read_csv('worldcities.csv')

    for i in inclusao:
        nome = i
        lat = df.loc[(df['city']==i)&(df['country']=='Germany'),'lat'].to_list()[0]
        lng = df.loc[(df['city']==i)&(df['country']=='Germany'),'lng'].to_list()[0]
        aux = {'nome':nome,'lat':lat,'lng':lng}
        
        folium.Circle(
                    location=[aux['lat'], aux['lng']],
                    radius=km,  # 100 km em metros (1 km = 1000 metros)
                    popup='Círculo de 100 km',
                    color=color,  # Cor da linha do círculo
                    fill=True,  # Preenchimento do círculo
                    fill_opacity=0.05,
                    opacity=.2 # Opacidade do preenchimento
                    ).add_to(m)

def adiciona_linha(m,inclusao,color):
    
    pontos = []
    for i in inclusao:
        lat = df.loc[(df['city']==i)&(df['country']=='Germany'),'lat'].to_list()[0]
        lng = df.loc[(df['city']==i)&(df['country']=='Germany'),'lng'].to_list()[0]
        pontos.append([lat,lng])
        
    folium.PolyLine(pontos,
                    color = color,
                    dash_array=5,
                    opacity=.85,
                    tooltip='Semana 4',).add_to(m)
        
```
