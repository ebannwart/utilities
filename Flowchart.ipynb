# Flowchart creation example

```python
from schemdraw import flow
import schemdraw
```

```python
d = schemdraw.Drawing(fontsize=8)

d += (b := flow.Start(w=5, h=1).label('First day of work'))
d += flow.Arrow().down().length(d.unit/3)

d += (b01 := flow.Box(w=4, h=1).label('Identifies the rack to be useda'))
d += flow.Arrow().down().length(d.unit/3)

d += (b1 := flow.Box(w=4, h=1).label('Clean the sample rack'))
d += flow.Arrow().down().length(d.unit/3)

d += (b2 := flow.Box(w=4, h=1).label('Coil production'))
d += flow.Arrow().down().length(d.unit/3)

d += (d1 := flow.Decision(w=5, h=3, W='No', E='Yes',fontsize=8)
      .label('Inspection or leader requested to\nkeep a sample?'))

d += flow.Line().left().at(d1.W).length(d.unit/2)
d += flow.Line().up().toy(b2.W)
d += flow.Arrow().right().tox(b2.W)

d += flow.Arrow().right().at(d1.E).length(d.unit/3)
d += (b3 := flow.Box(w=5, h=1).anchor('W').label('Take a sample with the problem\nTake a photo of the sample'))

d += flow.Arrow().up().at(b3.N).length(d.unit/2+0.5)
d += (b4 := flow.Box(w=5, h=1).anchor('S').label('Sample storage on rack'))

d += flow.Arrow().left().at(b4.W).length(d.unit/2)


d.draw()

ffname = r'I:\Corporacao\GPC\04- LPC\Python\Dados\RackEntradaLPC.png'  # full filename
d.save(ffname) 

d.draw()
```
