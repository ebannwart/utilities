```python
from schemdraw import flow
import schemdraw

%matplotlib inline  
```

# wbs

```python
h = 1
w = 4
tam = 5

t_pai = 'Loja de Bicicleta'
t_filhos = ['Planejamento','Locação','Compra de Produtos','Operação']

t_netoss = [['Local Instalação','Análise Concorrentes','Plano de Negócio'],
          ['Contrato Locação','Melhorias'],
          ['Negociação\nFornecedores','Estratégia\nArmazenamento','Execução Compra'],
          ['Fluxo de Caixa','Treinamento','Divulgação','Indicadores\nDesempenho'],
       ]
```

```python
d = schemdraw.Drawing(fontsize=10)

d += (pai := flow.Box(w=w, h=h).label(t_pai))

nfilhos = (len(t_filhos) - 1)/2
for i,filho in enumerate(t_filhos):
    posicao = nfilhos - i
    if posicao>=0:
        d += flow.Line().down().at(pai.S).length(tam/8)
        d += flow.Line().left().length(posicao*tam)
        d += flow.Line().down().length(tam/8)
        d += (f := flow.Box(w=w, h=h).label(filho))

        for ineto,neto in enumerate(t_netoss[i]):
            d += flow.Line().down().at(f.S).length((ineto+1)*tam/4)
            d += flow.Line().right().length(tam/8)
            d += (n := flow.Box(w=w, h=h*0.9).anchor('W').label(neto))

    else:
        posicao *= -1
        d += flow.Line().down().at(pai.S).length(tam/8)
        d += flow.Line().right().length(posicao*tam)
        d += flow.Line().down().length(tam/8)
        d += (f := flow.Box(w=w, h=h).label(filho))           

        for ineto,neto in enumerate(t_netoss[i]):
            d += flow.Line().down().at(f.S).length((ineto+1)*tam/4)
            d += flow.Line().right().length(tam/8)
            d += (n := flow.Box(w=w, h=h*0.8).anchor('W').label(neto))

        
d.draw()

ffname = r'ead.png'  # full filename
d.save(ffname,dpi=300) 

d.draw()
```
